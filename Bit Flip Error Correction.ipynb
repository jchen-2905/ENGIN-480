```python
from qiskit import QuantumCircuit, transpile
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Create a 3-qubit quantum circuit
qc = QuantumCircuit(3)

# Apply single-qubit gates
qc.h(0)  # Hadamard gate on qubit 0
qc.x(1)  # Pauli-X gate on qubit 1
qc.y(2)  # Pauli-Y gate on qubit 2
qc.z(0)  # Pauli-Z gate on qubit 0
qc.s(1)  # S (phase) gate on qubit 1
qc.t(2)  # T (pi/8) gate on qubit 2

# Apply two-qubit gates
qc.cx(0, 1)  # CNOT gate (control qubit 0, target qubit 1)
qc.cz(1, 2)  # Controlled-Z gate (control qubit 1, target qubit 2)
qc.swap(0, 2)  # SWAP gate between qubits 0 and 2

# Apply three-qubit gates
qc.ccx(0, 1, 2)  # Toffoli (CCX) gate
qc.cswap(0, 1, 2)  # (CSWAP) gate

# Apply rotation gates
qc.rx(0.5, 0)  # Rotation around X-axis on qubit 0
qc.ry(1.0, 1)  # Rotation around Y-axis on qubit 1
qc.rz(1.5, 2)  # Rotation around Z-axis on qubit 2

# Apply an arbitrary unitary matrix to qubit 0
from qiskit.circuit.library import U3Gate
qc.append(U3Gate(1.2, 2.3, 3.4), [0])

# Measure all qubits
qc.measure_all()
plt.show()
```
